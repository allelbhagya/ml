```python
import numpy as np

X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0,0,0,1])
```

```python
weights = np.array([0.3, 0.2])
bias = 1
eta = 0.1
```

```python
def f(weights, input, bias):
    z = np.dot(weights, input) + bias
    return 1 if z>0 else 0
```

```python
for j in range(10):
    for i in range(len(X)):
        predicted = f(X[i], weights, bias)
        weights += eta * (y[i] - predicted) *  X[i]
        bias += eta * (y[i] - predicted)
```

```python
print(weights, bias)
```

```python
for i in range(len(X)):
    predicted = f(X[i], weights, bias)
    print(X[i], predicted, y[i])
```
