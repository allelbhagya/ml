```python
test = [156,45]
k = 5
```

```python
import pandas as pd

df = pd.read_csv('./data.csv')
df
```

```python
dist = {}
for index,rows in df.iterrows():
    print(rows['height'], rows['weight'])
    dist[index] = ( (rows['height'] - test[0])**2 + (rows['weight'] - test[1])**2 ) ** 0.5
dist
```

```python
desc_val = dist.values()
desc_val = list(desc_val)
sorted_desc = sorted(desc_val, reverse=True)
sorted_desc = sorted_desc[:k]
print(sorted_desc)
```

```python
ind_k = []

for i in sorted_desc:
    for j in dist:
        if dist[j] == i:
            ind_k.append(j)
ind_k
```

```python
y_vals = []

for index in ind_k:
    for j,rows in df.iterrows():
        if j == index:
            print(rows['healthy'])
            y_vals.append(rows['healthy'])
```

```python
k_0 = k_1 = 0

for i in y_vals:
    if i == 0:
        k_0+=1
    else:
        k_1+=1
print(k_0,k_1)
if k_0>k_1:
    print("test belongs to class 0")
else:
    print("test belongs to class 1")
```
